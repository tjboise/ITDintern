```python
import pandas as pd
from datetime import datetime, timedelta
import requests
import os
```

```python
data = pd.read_excel('../download_images/idaho DOT.xlsx')
data['Station ID']
```

```python
end_time = datetime.now()-timedelta(minutes=15)
start_time = current_time - timedelta(days=1)

# start_time = datetime(2023, 6, 20, 0, 0, 0)  # Start time of the day
# end_time = datetime(2023, 6, 20, 23, 59, 45)  # End time of the day

current_time = start_time
delta = timedelta(minutes=15)
time_list=[]

while current_time <= end_time:
#     print(current_time.strftime('%Y-%m-%d %H:%M:%S'))
    time_list.append(current_time.strftime('%Y-%m-%d %H:%M:%S'))
    current_time += delta
# print(time_list)
```

```python
#get multiple image
date = '620'
for siteid in data['Station ID']:

    folder_path = '../../Github/ITDintern/download_images/'+'day'+date+'-siteid-'+str(siteid)+'/'  # Replace with the desired folder path

    print(folder_path)
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Folder '{folder_path}' created.")
    else:
        print(f"Folder '{folder_path}' already exists.")

    for timereq in time_list:

        image_url = f'https://exportdb.vaisala.io/export/image.jpg?username=idt&password=Data4Idaho&station={siteid}&cam=1&latesttime={timereq}'

        img_data = requests.get(image_url).content

        with open(folder_path+'/imagetemp_{}_{}.jpg'.format(siteid, timereq).replace(':', ''), 'wb') as handler:

            handler.write(img_data)
#         print(timereq)
```
